### Why do we need pandas in Python?

We need the pandas library in Python because it provides powerful, flexible, and **easy-to-use data structures**, primarily the DataFrame, which are essential for efficient data manipulation, analysis, and cleaning of structured (tabular) data. It simplifies complex tasks that would be cumbersome to perform using only built-in Python features like lists or dictionaries. [1, 2, 3, 4]  
Key Reasons for Using Pandas: 

• Efficient Data Structures: Pandas introduces  (one-dimensional labeled arrays) and  (two-dimensional tabular structures akin to spreadsheets or SQL tables) that are optimized for performance and memory usage, unlike general-purpose Python containers. 
• Data Cleaning and Wrangling: A significant portion of a data scientist's time is spent cleaning messy, real-world data. Pandas streamlines common tasks like: 

	• Handling missing data (e.g., detecting, removing, or filling  values). 
	• Removing duplicates. 
	• Renaming columns and changing data types. 
	• Transforming and reshaping datasets (e.g., pivoting, merging, joining, concatenating). 

• Data Analysis and Manipulation: Pandas offers extensive functionality for analyzing data and deriving insights: 

	• Filtering and Subsetting: Easily select specific rows and columns based on labels or conditions. 
	• Aggregation and Grouping: Perform "split-apply-combine" operations to group data by categories and compute summary statistics like mean, median, and sum. 
	• Time-Series Functionality: Specialized tools for working with date and time data, including date range generation and moving window calculations, crucial for finance and other fields.

### Series vs DataFrame

###  Series

A **Series** is:

* 1D data structure
* Like a NumPy array
* Has an index (labels)

Example structure mentally:

```
Index   Value
0       22
1       35
2       40
```

Shape → `(n,)`

You get a Series when you select a single column:

```python
df["Age"]
```

Returns → Series
Shape → `(n_samples,)`

---

### DataFrame

A **DataFrame** is:

* 2D tabular data
* Collection of multiple Series
* Each column is a Series

Shape → `(rows, columns)`

Selecting multiple columns:

```python
df[["Age", "Fare"]]
```

Returns → DataFrame
Shape → `(n_samples, 2)`

---

### Important Dimensional Rule

| Syntax            | Returns   | Shape   |
| ----------------- | --------- | ------- |
| `df["col"]`       | Series    | `(n,)`  |
| `df[["col"]]`     | DataFrame | `(n,1)` |
| `df[["c1","c2"]]` | DataFrame | `(n,2)` |

Why?

Because passing a **list of column labels** means you are selecting a table subset → 2D.

### Inspecting Data

Proper inspection order:

### 1. `df.shape`

Returns:

```
(rows, columns)
```

Example:

```
(366, 22)
```

Means:

* 366 samples
* 22 columns

Not all columns are features — one may be target.

---

### 2. `df.columns`

Shows column labels.

Important for:

* Selecting features
* Identifying target
* Avoiding spelling mistakes

---

### 3. `df.head()`

Shows first 5 rows.

Used to visually inspect structure.

---

### 4. `df.info()`

Shows:

* Number of rows
* Column names
* Non-null count
* Data types

Critical for detecting:

* Wrong dtype
* Missing data
* Object columns that should be numeric

---

### 5. `df.describe()`

By default:

```python
df.describe()
```

Returns summary statistics for numeric columns only:

* count
* mean
* std
* min
* 25%
* 50%
* 75%
* max

If you want all columns:

```python

```

For object columns, it shows:

* count
* unique
* top (mode)
* freq

## Row Selection — loc vs iloc

### loc → Label-based

Selects using index labels.

If index =

```
100, 101, 102
```

Then:

```python
df.loc[100]
```

Selects row with label 100.

If label doesn't exist → KeyError.


#### Slicing with loc

```python
df.loc[100:102]
```

End is INCLUDED.

---

## iloc → Position-based

Selects using position (0-based).

```python
df.iloc[0]
```

Selects first row (position 0).

Independent of label values.

### Slicing with iloc

```python
df.iloc[0:3]
```

End is EXCLUDED.

---

## 🔥 Important Difference

|           | loc       | iloc      |
| --------- | --------- | --------- |
| Based on  | Labels    | Positions |
| Slice end | Inclusive | Exclusive |

---

````python
# Check labels for rows
print(df.index)

# By default for rows indices start from 0
# indices are always integer starting from 0 to total rows - 1
# labels are names they can anything string, number etc

# Check labels for the columns
print(df.columns)
````

````python
# Column Selection

df["Age"] #  returns Series
df[["Age"]] # returns data frame
````

````python
# Row selection

total_rows, *rest = df.shape
labels = [100 + x for x in range(0,total_rows)]

df.index = labels
print(df)

# you can see the change in row labels
````

````python
# Label Based Row Selection (loc)

print(df.loc[100]) # this is label not index
print(df.loc[100:200]) # slicing works here

# Column selection with loc
print(df.loc[:,"Age"])
````

````python
# Position based row selection
# indices zeros based so range is 0 to total_rows - 1

print(df.iloc[0])
print(df.iloc[0:33])

# similarly you can indices for columns too
print(df.iloc[:,0:4])
````

### View vs Copy 

Pandas may return:

* View
* Copy
* Ambiguous object

You should not rely on view behavior.

If modifying sliced data:

```python
df2 = df[["col"]].copy()
```

Prevents unintended side effects.

### Boolean Filtering 

### Step 1: Create mask

```python
df["MinTemp"] > 10
```

Returns:

* Series of True/False
* Same length as DataFrame

Example:

```
True
False
True
...
```

---

### Step 2: Apply mask

```python
df[df["MinTemp"] > 10]
```

Returns:

* Filtered rows
* All columns preserved

If 100 rows match:

New shape:

```
(100, original_columns)
```

---

### Combining Conditions

Use:

* `&` → AND
* `|` → OR
* `~` → NOT

Never use:

* `and`
* `or`

---

## Correct Structure

```python
(df["MinTemp"] > 10) & (df["MaxTemp"] < 25)
```

Parentheses required because:

* `&` has higher precedence than `>`

Without parentheses → wrong evaluation.
