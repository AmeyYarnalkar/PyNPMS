```python
import pandas as pd
import numpy as np

data = {
    "OrderID": range(1, 16),
    "Category": [
        "Electronics", "Clothing", "Electronics", "Home",
        "Clothing", "Electronics", "Home", "Clothing",
        "Electronics", "Home", "Clothing", "Clothing",
        "Electronics", np.nan, "Home"
    ],
    "PaymentMethod": [
        "UPI", "Card", "UPI", "Cash",
        "Card", "UPI", "Cash", "Card",
        "Card", "Cash", "UPI", "Card",
        "UPI", "UPI", "Cash"
    ],
    "Amount": [
        15000, 2000, 12000, 5000,
        2500, 18000, 7000, 3000,
        22000, 4500, 2800, 3500,
        16000, 9000, 6000
    ]
}

df = pd.DataFrame(data)
df
```

```python
df.sort_values("Amount")
```

```python
df.sort_values("Amount", ascending=False)
```

```python
df.sort_values("Category")
```

```python
df.sort_values(["Category", "Amount"], ascending=[True, False])
```

```python
df.sort_values("Category", na_position="first") # it can be last also
```

```python
# Sorting keeps the row index same as original if you want reset it as per new sorting then
df.sort_values("Amount").reset_index(drop=True)
```

```python
df.sort_index()
```

```python
df.sort_index(axis=1)
```
