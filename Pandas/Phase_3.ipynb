````python
# Create a data set
import pandas as pd

data = {
    "user_id": [1,2,3,4,5,6],
    "age": [22, 35, 28, 45, 31, 52],
    "gender": ["Male","Female","male","Female","Mle","Male"],
    "city": ["Mumbai","Delhi","Mumbai","Bangalore","Delhi","Mumbai"],
    "signup_date": [
        "2024-01-01",
        "2024-02-15",
        "2024-02-20",
        "2024-03-05",
        "2024-03-18",
        "2024-04-01"
    ],
    "salary": [40000,60000,52000,80000,58000,120000]
}

df = pd.DataFrame(data)

df["signup_date"] = pd.to_datetime(df["signup_date"])

df
````

#### Phase 3 : Feature Engineering Basics

#### Phase 3.1 — Creating New Columns

This is the MOST important habit.

#### Example 1 — Interaction Feature

```python
df["salary_per_age"] = df["salary"] / df["age"]
```

Why?

Models often learn better from ratios than raw numbers.

---

#### Example 2 — Date Features

```python
df["signup_month"] = df["signup_date"].dt.month
df["signup_dayofweek"] = df["signup_date"].dt.dayofweek
```

---

#### Phase 3.2 — `map()` and `replace()` (Preferred over apply)

You already saw messy gender values:

```
Male, male, Mle
```

Normalize first.

```python
df["gender"] = df["gender"].str.lower()

df["gender"] = df["gender"].replace({
    "mle": "male"
})
```

Now convert categories into numbers:

```python
df["gender_encoded"] = df["gender"].map({
    "male": 1,
    "female": 0
})
```

Why `map()`?

👉 Fast
👉 Clean
👉 Vectorized

---

#### Phase 3.3 — Controlled use of `apply()`

Use only when vectorized logic doesn’t exist.

Example: classify salary level.

```python
def salary_level(x):
    if x < 50000:
        return "low"
    elif x < 80000:
        return "medium"
    else:
        return "high"

df["salary_level"] = df["salary"].apply(salary_level)
```

But remember:

👉 If pandas has built-in methods, prefer them.

---

#### Phase 3.4 — One-Hot Encoding (Just Know this we will learn it soon)

Convert categories into ML-friendly format.

```python
df = pd.get_dummies(df, columns=["city"], dtype=int)
```

Now you’ll see:

```
city_Mumbai
city_Delhi
city_Bangalore
```

This is extremely common before training models.

---

#### Phase 3.5 — Binning (cut / qcut)

Turning continuous numbers into groups.

#### Equal-width bins

```python
df["age_group"] = pd.cut(
    df["age"],
    bins=[0,30,50,100],
    labels=["young","adult","senior"]
)
```

---

#### Quantile bins (balanced groups)

```python
df["salary_quantile"] = pd.qcut(df["salary"], q=3)
```

This creates bins with roughly equal counts.

Very useful when distributions are skewed.

---
