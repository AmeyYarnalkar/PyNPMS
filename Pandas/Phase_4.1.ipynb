````python
import pandas as pd
import numpy as np

data = {
    "OrderID": range(1, 16),
    "Category": [
        "Electronics", "Clothing", "Electronics", "Home",
        "Clothing", "Electronics", "Home", "Clothing",
        "Electronics", "Home", "Clothing", "Clothing",
        "Electronics", np.nan, "Home"
    ],
    "PaymentMethod": [
        "UPI", "Card", "UPI", "Cash",
        "Card", "UPI", "Cash", "Card",
        "Card", "Cash", "UPI", "Card",
        "UPI", "UPI", "Cash"
    ],
    "Amount": [
        15000, 2000, 12000, 5000,
        2500, 18000, 7000, 3000,
        22000, 4500, 2800, 3500,
        16000, 9000, 6000
    ]
}

df = pd.DataFrame(data)
df
````

### Category Analysis & Frequency Handling

#### What Is Category Analysis?

In ML, before modeling, you must answer:

* Which category appears most?
* Is there class imbalance?
* Are there rare categories?
* Are there missing values?
* How are numerical values distributed?

If you skip this → your model assumptions are weak.

---

#### value_counts() — Core Frequency Tool

#### Basic Frequency Count

```python
df["Category"].value_counts()
```

👉 This tells you:

* Electronics: ?
* Clothing: ?
* Home: ?

It sorts automatically in descending order.

This is default behavior.

---

#### value_counts(normalize=True)

```python
df["Category"].value_counts(normalize=True)
```

Now you get proportions instead of counts.

Why this matters:

In classification problems, you care about distribution.

Example:
If one category is 80% of data → imbalance problem.

This is how you detect it early.

---

#### Handling Missing Values in Frequency

By default, NaN is ignored.

```python
df["Category"].value_counts(dropna=False)
```

Now you’ll see NaN count.

---

#### value_counts() on Numerical Data

You can also bin numbers.

```python
df["Amount"].value_counts(bins=3)
```

This creates ranges like:

(2000–9000], (9000–16000], etc.

Why useful?

Quick way to:

* See spread
* Detect skew
* Roughly segment customers

But don’t use this blindly — bins are arbitrary unless chosen properly.

---

#### unique() vs nunique()

### unique()

```python
df["PaymentMethod"].unique()
```

Returns actual values.

#### nunique()

```python
df["PaymentMethod"].nunique()
```

Returns count of unique categories.

Why this matters?

If categorical column has 5000 unique values →
One-hot encoding will explode.

You must check this before feature engineering.

---

#### Sorting Frequency Results

You can control order:

```python
df["Category"].value_counts().sort_index()
```

Sort alphabetically.

Or:

```python
df["Category"].value_counts().sort_values(ascending=True)
```

Sort by frequency ascending.

Useful when analyzing rare categories.

---

#### Combining With Filtering

Example:

High value orders only:

```python
df[df["Amount"] > 10000]["Category"].value_counts()
```

Now you're doing conditional frequency analysis.

This is powerful in EDA.

````python
# Conditional filtering
df[df["Amount"] > 10000]["Category"].value_counts()

""" 
How is it evaluated?
1. df["Amount"] > 10000 gives us mask for this condition
2. df[df["Amount"] > 10000] returns a dataframe containing rows that result true for the provided condition
3. out of that data frame we want only "Category" column and their value_count()
"""
````
