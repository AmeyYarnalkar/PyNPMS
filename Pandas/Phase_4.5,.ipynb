### Crosstab 

#### What is Crosstab?

Crosstab = Frequency table between two (or more) categorical variables.

It answers questions like:

* Do males complete ML more than females?
* Is completion dependent on course?
* Is there imbalance?

Think of it as a contingency table.

---

#### Basic Crosstab (Two Variables)

```python
pd.crosstab(df["Course"], df["Completed"])
```

This gives:

Rows → Course
Columns → Completed
Values → Count of occurrences

It counts how many ML students completed vs not completed.

Behind the scenes:

It groups by both columns and counts frequency.

Equivalent mental model:

GroupBy on two columns → count rows → reshape table.

---

#### Crosstab with Multiple Variables

You can cross more than one variable.

Example: Course vs Gender

```python
pd.crosstab(df["Course"], df["Gender"])
```

Now you see gender distribution per course.

---

You can even do multi-level columns:

```python
pd.crosstab(df["Course"], [df["Gender"], df["Completed"]])
```

Now columns become hierarchical:

Gender → Completed

This is very powerful for segmentation.

---

#### normalize= (Critical for ML)

Raw counts can mislead.

Use normalize to see proportions.

### Normalize by row

```python
pd.crosstab(df["Course"], df["Completed"], normalize="index")
```

Each row sums to 1.

Meaning:
Completion rate within each course.

This is usually what you want.

---

#### Normalize by column

```python
pd.crosstab(df["Course"], df["Completed"], normalize="columns")
```

Each column sums to 1.

Meaning:
Among those who completed, how many were from each course.

Different interpretation.

---

#### Normalize entire table

```python
pd.crosstab(df["Course"], df["Completed"], normalize=True)
```

Entire table sums to 1.

---

#### Adding Margins (Totals)

```python
pd.crosstab(df["Course"], df["Completed"], margins=True)
```

Adds:

* Row totals
* Column totals
* Grand total

Good for sanity checking counts.

---

#### Combine normalize + margins

```python
pd.crosstab(
    df["Course"],
    df["Completed"],
    normalize="index",
    margins=True
)
```
