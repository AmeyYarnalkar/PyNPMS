### Step 3 — Descriptive Statistics (Critical Step)

Run:

```python

```

This gives you:

* Count
* Mean
* Standard deviation
* Minimum
* 25th percentile (Q1)
* 50th percentile (Median)
* 75th percentile (Q3)

Now don’t just stare at numbers. Interpret them.

---

### 1️ Mean vs Median — Detect Skewness

Compare:

* **Mean**
* **50% (Median)**

### If:

* Mean > Median → Right skewed
* Mean < Median → Left skewed

In video game sales data, most sales columns are **right skewed**.

Why?

Because:

* Most games sell small numbers
* A few blockbuster games sell extremely high

That creates a long right tail.

Example:

* Mean = 0.5 million
* Median = 0.15 million
* Max = 82 million

That’s heavy right skew.

This tells you:
Most games sell very little. A few dominate the market.

That’s real insight.

---

### Minimum and Maximum — Reality Check

Look at:

* Minimum
* Maximum

Ask yourself:

* Is the minimum reasonable?
* Is the maximum extremely large compared to mean?

If:

* Max is 100x bigger than median
* Or standard deviation is very high

Then the distribution is spread out heavily.

That signals outliers or extreme variance.

---

### 3️ Standard Deviation — Spread of Data

Standard deviation tells you how much values deviate from the mean.

If standard deviation is large compared to the mean:

That means data is highly dispersed.

In sales datasets, this is common.

Why?

Because:

* Most games cluster near small values
* A few explode in sales

High spread = high variability in performance.

---

### 4️ Are Most Values Near Zero?

Even before plotting, think:

* Is Q1 very close to 0?
* Is median small?
* Is max extremely high?

If yes:

Most data is clustered near lower values, and only a few are very high.

That’s classic long-tail distribution.

---

### Outliers Are Not Always Errors

In video game sales:

Blockbusters are expected outliers.

Examples of legitimate extreme sellers:

* Grand Theft Auto V
* Super Mario Bros.
* Wii Sports

These are not data mistakes.
They are real-world dominance.

Later, when you plot a boxplot, these will show up as extreme points.

Don’t panic. That’s expected.

---

### What You Should Conclude After `describe()`

You should be able to say something like:

* “Global sales are highly right skewed.”
* “Most games sell under 1 million.”
* “A few games dominate total sales.”
* “There is high variability in performance.”

If you can say that confidently, you are thinking like a data analyst.

If you’re just reading numbers without conclusions, you’re wasting time.
